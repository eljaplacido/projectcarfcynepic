# CARF Quickstart

This notebook demonstrates how to use the CARF cognitive architecture
for enterprise decision support. You'll learn to:

1. Classify queries into Cynefin domains
2. Run the full CARF pipeline
3. Inspect results and reasoning chains
4. Export reproducibility artifacts

```python
import sys
sys.path.insert(0, '..')

from src.api.library import (
    classify_query,
    run_pipeline,
    get_system_capabilities,
    export_reproducibility_artifact,
)
```

## 1. System Capabilities

Discover what CARF can do:

```python
caps = get_system_capabilities()
print('Cynefin Domains:')
for d in caps['cynefin_domains']:
    print(f"  {d['domain']}: {d['description']}")

print(f"\nAvailable benchmarks: {len(caps['benchmarks'])}")
print(f"Library functions: {caps['library_functions']}")
```

## 2. Classify a Query

The CARF router classifies queries into Cynefin domains:

```python
result = await classify_query("What is the causal effect of discounts on customer churn?")
print(f"Domain: {result['domain']}")
print(f"Confidence: {result['confidence']:.2%}")
print(f"Entropy: {result['entropy']:.4f}")
```

## 3. Run the Full Pipeline

Execute the complete CARF cognitive pipeline (Router → Domain Agent → Guardian):

```python
pipeline_result = await run_pipeline(
    "Analyze why our supplier costs increased 15% last quarter"
)

print(f"Domain: {pipeline_result['domain']}")
print(f"Confidence: {pipeline_result['domain_confidence']:.2%}")
print(f"Reasoning steps: {pipeline_result['reasoning_steps']}")
print(f"\nResponse:\n{pipeline_result['response'][:500]}")
```

## 4. Export Reproducibility Artifact

Package results for audit and reproducibility:

```python
artifact = export_reproducibility_artifact(pipeline_result)
print(f"Artifact type: {artifact['artifact_type']}")
print(f"Exported at: {artifact['exported_at']}")
print(f"Python: {artifact['environment']['python_version'][:20]}...")
```
