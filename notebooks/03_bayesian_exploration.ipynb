# CARF Bayesian Exploration

This notebook demonstrates CARF's Bayesian active inference capabilities:

1. Update beliefs with observational data
2. Quantify epistemic vs aleatoric uncertainty
3. Design probes to reduce uncertainty
4. Explore energy demand forecasting

```python
import sys
sys.path.insert(0, '..')

from src.api.library import run_bayesian, classify_query, export_reproducibility_artifact
```

## 1. Classify as Complex Domain

Bayesian queries typically map to the Complex Cynefin domain:

```python
classification = await classify_query(
    "How uncertain is our conversion rate estimate given recent A/B test data?"
)
print(f"Domain: {classification['domain']}")
print(f"Confidence: {classification['confidence']:.2%}")
```

## 2. Update Beliefs with Observations

Run Bayesian inference to update our belief about conversion rates:

```python
# Monthly conversion observations
observations = [0.42, 0.38, 0.45, 0.41, 0.39, 0.44, 0.40, 0.43]

result = await run_bayesian(
    query="What is our true conversion rate given these observations?",
    observations=observations,
)

print(f"Posterior mean: {result.get('posterior_mean', 'N/A')}")
print(f"Epistemic uncertainty: {result.get('epistemic_uncertainty', 'N/A')}")
print(f"Aleatoric uncertainty: {result.get('aleatoric_uncertainty', 'N/A')}")
print(f"\nResponse:\n{result.get('response', 'N/A')[:500]}")
```

## 3. Energy Demand Forecasting

Use Bayesian inference for time-series forecasting:

```python
# Monthly energy demand in MW
energy_demand = [1200, 1350, 1180, 1420, 1300, 1480, 1220, 1390, 1250, 1450, 1320, 1500]

forecast = await run_bayesian(
    query="Forecast energy demand for next quarter considering seasonal patterns",
    observations=energy_demand,
)

print(f"Forecast mean: {forecast.get('posterior_mean', 'N/A')}")
print(f"Epistemic uncertainty: {forecast.get('epistemic_uncertainty', 'N/A')}")
```
